```python
import plotly.graph_objects as go
import numpy as np

x = [0, 2, 4, 5, 7, 8, 9]
y = [0, 0.003, 0.01, 0.03, 0.06, 0.12, 0.11]
cats = [[2], [2, 4], [4], [5], [2,3,4,5], [1]]
possible_cats = 5
n_points = 1000
x_range = max(x) - min(x)
t = np.linspace(min(x), max(x), n_points)
xx = []
yy = []
cc = []
for idx, ((x0, x1), (y0, y1)) in enumerate(zip(zip(x[:-1], x[1:]), zip(y[:-1], y[1:]))):
    print(((x0, x1), (y0, y1)))
    prop = (x1 - x0)/x_range
    points = int(n_points*prop)
    xx.extend(np.linspace(x0, x1, points))
    yy.extend(np.linspace(y0, y1, points))
    repeats = int(points/len(cats[idx]))
    colors = [val for val in list(sorted(cats[idx])) for _ in range(repeats)]
    colors = colors + [cats[idx][-1]*(points - len(colors))]
    cc.extend([float(x-1)/(possible_cats-1) for x in colors])

data = go.Scatter(x=xx, y=yy, mode='markers', marker={'color': cc, 'colorscale': 'Rainbow', 'size': 10})

layout = dict(plot_bgcolor='white', margin=dict(t=0, b=0, r=0, l=0, pad=0),
              xaxis=dict(showgrid=False, zeroline=False, mirror=True, linecolor='gray'),
              yaxis=dict(showgrid=False, zeroline=False, mirror=True, linecolor='gray'))

fig = go.Figure(data=data, layout=layout)

fig.show()
```
